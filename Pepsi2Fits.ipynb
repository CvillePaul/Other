```python
from glob import glob
import os
from astropy.io import fits
import numpy as np

spectra = []
for path in glob("../Files/2023??/pepsi*.*"):
    hdul = fits.open(path)
    hdr = hdul[0].header
    data = np.array([hdul[1].data["Arg"], hdul[1].data["Fun"]])

    hdr["HISTORY"] = f"Converted to IRAF-compatible format by Pepsi2Fits.ipynb, by Paul McKee"

    newhdu = fits.PrimaryHDU(header=hdr, data=data)
    new_filename = os.path.basename(path).replace(".nor", ".fits").replace(".bwl", ".fits")
    newhdu.writeto(new_filename, overwrite=True)
    # print(path, new_filename)
```
